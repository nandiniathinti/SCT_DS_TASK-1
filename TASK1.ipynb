```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set_style("whitegrid")
```

```python
df = pd.read_csv(
    "dataset/API_SP.POP.TOTL_DS2_en_csv_v2_451027.csv",
    skiprows=4
)

df.head()
```

```python
print("Rows and Columns:", df.shape)
```

```python
print(df.columns)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
population = df[['Country Name', '2022']]
population.head()
```

```python
population = population.dropna()
```

```python
population.isnull().sum()
```

```python
population = population.sort_values(by='2022', ascending=False)
population.head()
```

```python
top10 = population.head(10)

top10
```

```python
plt.figure(figsize=(12,6))

sns.barplot(
    data=top10,
    x='Country Name',
    y='2022',
    palette='viridis'
)

plt.title("Top 10 Countries by Population (2022)", fontsize=15)

plt.xlabel("Country")

plt.ylabel("Population")

plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(10,6))

plt.hist(population['2022'], bins=20, edgecolor='black')

plt.title("Distribution of Population")

plt.xlabel("Population")

plt.ylabel("Frequency")

plt.show()
```
